```python
df.Category = df.Category.replace('1.9' , np.nan) 
df.Category = df.Category.str.replace('_' , ' ')
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'],errors='coerce')
```

```python
df.Size = df.Size.apply(lambda x : str(x).replace('Varies with device' , '0'))

filter_k =  df.Size.apply(lambda x : 'k' in x)
df.Size[filter_k] = df.Size[filter_k].apply(lambda x : str(np.double(str(x).replace('k' , ''))/1000))

df.Size = df.Size.apply(lambda x : str(x).replace('M' , ''))
df.Size = pd.to_numeric(df.Size , errors="coerce")
```

```python
x=[True,False,False,False,False,True,False,True,False,True,False,False,True]
df.iloc[:,x]
```

```python
df.Type = df.Type.replace('0' , np.nan) 
df.Installs = df.Installs.apply(lambda x: re.sub("[+,]", "",x))
```

## drop NaN rows

```python
df.dropna(inplace = True)
```

## convert price to float

## Converting Content Rating to 0 and 1 with one hot encoding


```python
content_rating = pd.get_dummies(df['Content Rating'])
df = pd.concat([df,content_rating] , axis=1)
```

```python
df.head(10)
```

```python
correlation_df = df.corr()
```

```python
sns.heatmap(correlation_df,cmap='PiYG' , annot=True)
```

## Converting  Type to 0 and 1 with one hot encoding

```python
Types = pd.get_dummies(df.Type)
df = pd.concat([df,Types],axis=1)
```

## replace $ 

```python
df.Price = df.Price.apply(lambda x : str(x).replace('$' , ''))
df.Price = pd.to_numeric(df.Price , errors="coerce" )
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = df.loc[: , ['Size','Reviews' , 'Free', 'Paid']]
```

```python
Y=df.Installs
```

```python
X_train, X_test , y_train , y_test = train_test_split(X,Y , test_size=0.2)
```

```python
 regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
predictions=regressor.predict(X_test)
```
